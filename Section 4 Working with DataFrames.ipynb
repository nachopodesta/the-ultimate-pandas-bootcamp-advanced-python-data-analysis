```python
def convert_list_to_dict(l):
    return {k:v for k,v in enumerate(l)}
```

```python
dict_names = convert_list_to_dict(names)
dict_ages = convert_list_to_dict(ages)
dict_married = convert_list_to_dict(married)
```
